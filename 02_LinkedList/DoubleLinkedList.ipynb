# Double Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.previous = None
        self.next = None


class DoubleLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def __str__(self):
        temp_node = self.head
        out = ''

        if temp_node is None:
            out = "List is empty"
            return
          
        while temp_node:
            out += str(temp_node.data) + '->'
            temp_node = temp_node.next
        return out[:-2]

    def __len__(self):
        count = 0
        temp_node = self.head
        while temp_node:
            temp_node = temp_node.next
            count += 1

        return count

    def insertAfter(self, prev_data, new_data):
        temp_node = self.head
        while temp_node.next:
            if temp_node.data == prev_data:
                new_node = Node(new_data)
                new_node.next = temp_node.next
                temp_node.next.previous = new_node
                temp_node.next = new_node
                new_node.previous = temp_node
                return
            temp_node = temp_node.next

        print("The given previous node is not available in Linked List.")

    def addNodeAtHead(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = self.tail = new_node
        else:
            self.head.previous = new_node
            new_node.next = self.head
            self.head = new_node

    def addNodeAtTail(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.previous = self.tail
            self.tail = new_node

    def deleteNodeAtHead(self):
        if self.head:
            self.head = self.head.next
            self.head.previous = None
            return
        else:
            print("Linked List is empty")

    def deleteNodeAtTail(self):
        if self.tail:
            self.tail = self.tail.previous
            self.tail.next = None

    def deleteNode(self, data):
        temp_node = self.head

        if temp_node and temp_node.data == data:
            self.deleteNodeAtHead()
            return
        elif self.tail and self.tail.data == data:
            self.deleteNodeAtTail()
            return
        else:
            while temp_node:
                if temp_node.data == data:
                    previous_node.next = temp_node.next
                    temp_node.next.previous = previous_node
                    return
                previous_node = temp_node
                temp_node = temp_node.next
            print("The given node is not available in Linked List. ")
```

```python
dll = DoubleLinkedList()

dll.addNodeAtHead(1)
dll.addNodeAtTail(2)
dll.addNodeAtTail(3)
dll.insertAfter(2, 2.1)

print("DLL: " + str(dll))

dll.insertAfter(2.1, 2.5)

print("DLL: " + str(dll))
dll.addNodeAtHead(0)
print("DLL: " + str(dll))

dll.deleteNodeAtHead()
dll.deleteNodeAtTail()
print("DLL: " + str(dll))

dll.deleteNode(2.1)

print("DLL: " + str(dll))

print("Length of CLL: " + str(len(dll)))
```
