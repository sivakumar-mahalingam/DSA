# CIRCULAR LINKED LIST

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class CircularLinkedList:
    def __init__(self):
        self.head = None

    def __len__(self):
        if self.head is None:
            return 0

        count = 0
        temp = self.head
        while True:
            count += 1
            temp = temp.next
            if temp == self.head:
                break
        return count

    def __str__(self):
        temp_node = self.head
        out = ''

        if temp_node is None:
            out = "List is empty"
            return out
        elif temp_node is not None:
            while True:
                out += str(temp_node.data) + '->'
                temp_node = temp_node.next
                if temp_node == self.head:
                    return out[:-2]
    
    def insertAtBeginning(self, data): 
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            new_node.next = self.head
        else:
            temp_node = self.head
            while temp_node.next != self.head:
                temp_node = temp_node.next
            new_node.next = self.head
            temp_node.next = new_node
            self.head = new_node
            
    def insertNodeAfter(self, prev_data, new_data):
        temp_node = self.head
        while temp_node:
            if temp_node.data == prev_data:
                new_node = Node(new_data)
                new_node.next = temp_node.next
                temp_node.next = new_node
                return
            temp_node = temp_node.next

        print("The given previous node is not found in LinkedList.")

    def insertAtEnd(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            new_node.next = self.head
        else:
            temp_node = self.head
            while temp_node.next != self.head:
                temp_node = temp_node.next
            temp_node.next = new_node
            new_node.next = self.head
            
        def deleteAtBeginning(self):
        if self.head is None:
            print("List is empty")
            return

        if self.head.next == self.head:
            self.head = None
            return

        temp_node = self.head
        while temp_node.next != self.head:
            temp_node = temp_node.next
        temp_node.next = self.head.next
        self.head = self.head.next

    def deleteNode(self, data):
        if self.head is None:
            print("List is empty.")
            return

        if self.head.data == data:
            temp_node = self.head
            while temp_node.next != self.head:
                temp_node = temp_node.next
            temp_node.next = self.head.next
            self.head = self.head.next
            if self.head == self.head.next:
                self.head = None
            return

        temp_node = self.head
        prev = None
        while temp_node.next != self.head:
            prev = temp_node
            temp_node = temp_node.next
            if temp_node.data == data:
                prev.next = temp_node.next
                return

        print("Node not found in the list.")

    def deleteAtEnd(self):
        if self.head is None:
            print("List is empty")
            return

        if self.head.next == self.head:
            self.head = None
            return

        temp_node = self.head
        while temp_node.next.next != self.head:
            temp_node = temp_node.next
        temp_node.next = self.head
```
