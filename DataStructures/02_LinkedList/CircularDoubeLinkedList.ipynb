# CIRCULAR DOUBLE LINKED LIST

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


class CircularDoublyLinkedList:
    def __init__(self):
        self.head = None

    def __str__(self):
        if self.head is None:
            return "List is empty"

        temp_node = self.head
        out = ""
        while True:
            out += str(temp_node.data) + "->"
            temp_node = temp_node.next
            if temp_node == self.head:
                return out[:-2]

    def __len__(self):
        if self.head is None:
            return 0

        count = 0
        temp = self.head
        while True:
            count += 1
            temp = temp.next
            if temp == self.head:
                break
        return count

    def addNodeAtHead(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            new_node.next = new_node
            new_node.prev = new_node
        else:
            last_node = self.head.prev
            new_node.next = self.head
            new_node.prev = last_node
            self.head.prev = new_node
            last_node.next = new_node
            self.head = new_node

    def addNodeAtLast(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            new_node.next = new_node
            new_node.prev = new_node
        else:
            last_node = self.head.prev
            new_node.next = self.head
            new_node.prev = last_node
            self.head.prev = new_node
            last_node.next = new_node

    def deleteNode(self, data):
        if self.head is None:
            print("List is empty.")
            return

        temp_node = self.head
        if temp_node.data == data:
            if temp_node.next == self.head:
                self.head = None
            else:
                last_node = self.head.prev
                self.head = self.head.next
                last_node.next = self.head
                self.head.prev = last_node
            return

        temp_node = temp_node.next
        while temp_node != self.head:
            if temp_node.data == data:
                temp_node.prev.next = temp_node.next
                temp_node.next.prev = temp_node.prev
                return
            temp_node = temp_node.next

        print("Node not found in the list.")
```

```python
cdll = CircularDoublyLinkedList()
cdll.addNodeAtLast(10)
cdll.addNodeAtLast(20)
cdll.addNodeAtLast(30)
print("CDLL: " + str(cdll))
print("Length of CDLL: " + str(len(cdll)))

cdll.addNodeAtHead(5)
print("CDLL: " + str(cdll))
print("Length of CDLL: " + str(len(cdll)))

cdll.deleteNode(20)
print("CDLL: " + str(cdll))
print("Length of CDLL: " + str(len(cdll)))

cdll.deleteNode(10)
cdll.deleteNode(5)
cdll.deleteNode(30)
print("CDLL: " + str(cdll))
print("Length of CDLL: " + str(len(cdll)))
```
