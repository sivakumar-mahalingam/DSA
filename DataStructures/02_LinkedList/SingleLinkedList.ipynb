# SINGLE LINKED LIST

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    def __str__(self):
        temp_node = self.head
        out = ''

        if temp_node is None:
            out = "List is empty"
            return out
          
        while temp_node:
            out += str(temp_node.data) + '->'
            temp_node = temp_node.next
        return out[:-2]

    def __len__(self):
        count = 0
        temp_node = self.head
        while temp_node:
            temp_node = temp_node.next
            count += 1

        return count

    def addNodeAfter(self, prev_data, new_data):
        temp_node = self.head
        while temp_node.next:
            if temp_node.data == prev_data:
                new_node = Node(new_data)
                new_node.next = temp_node.next
                temp_node.next = new_node
                return
            temp_node = temp_node.next

        print("The given previous node is not found in LinkedList.")

    def addNode(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            return

        temp_node = self.head
        while temp_node.next:
            temp_node = temp_node.next
        temp_node.next = new_node

    def deleteNode(self, data):
        temp_node = self.head
        
        if temp_node is None:
            print("List is empty")
            return

        if temp_node.data == data:
            self.head = temp_node.next
            return

        while temp_node:
            if temp_node.data == data:
                prev.next = temp_node.next
                return
            prev = temp_node
            temp_node = temp_node.next
```

```python
ll = LinkedList()

ll.addNode(1)
ll.addNode(2)
ll.addNode(3)

print("LL: " + str(ll))

ll.addNodeAfter(2, 2.5)

print("LL: " + str(ll))

ll.deleteNode(2)

print("LL: " + str(ll))

print("Length of LL: " + str(len(ll)))


ll.deleteNode(1)
ll.deleteNode(2.5)
ll.deleteNode(3)
print("LL: " + str(ll))
print("Length of LL: " + str(len(ll)))
```
